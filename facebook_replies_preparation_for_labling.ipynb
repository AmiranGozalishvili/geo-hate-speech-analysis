```python
import pandas as pd

# List of CSV file names
file_names = ['dataset_facebook-comments-scraper_2023-11-07_07-57-07-032 (1).csv',
              'dataset_facebook-comments-scraper_2023-11-07_08-17-01-020.csv',
              'dataset_facebook-comments-scraper_2023-11-07_08-24-38-544.csv',
              'dataset_facebook-comments-scraper_2023-11-07_08-34-55-797.csv',
              'dataset_facebook-comments-scraper_2023-11-07_08-40-17-585.csv',
              'dataset_facebook-comments-scraper_2023-11-07_09-38-10-768.csv',
              'dataset_facebook-comments-scraper_2023-11-07_10-04-50-954.csv']  # Update with your file names

# Initialize an empty dictionary to store DataFrames
dataframes = {}

# Loop through the file names and read the CSV files
for file_name in file_names:
    # Construct the full path to the CSV file
    file_path = f'/content/{file_name}'  # Modify the path as needed

    # Read the CSV file into a DataFrame and store it in the dictionary
    dataframes[file_name] = pd.read_csv(file_path)

# Now, you have a dictionary where the keys are the file names, and the values are DataFrames containing the data from the CSV files.
# Concatenate the dataframes from the dictionary into a single dataframe
dataframes_list = list(dataframes.values())
combined_dataframe = pd.concat(dataframes_list, ignore_index=True)
combined_dataframe
```

```python
# Define a regex pattern to match English and Georgian letters
pattern = re.compile(r'[a-zA-Zა-ჰ]+')

filtered_row = []

for row in combined_dataframe:
  filtered_comment = ' '.join(pattern.findall(row))
  filtered_row.append(filtered_comment)
  print(filtered_comment)
```

```python
import pandas as pd
import re

# Define a regex pattern to match English and Georgian letters
pattern = re.compile(r'[a-zA-Zა-ჰ]+')

# Create a new column to store the filtered text
combined_dataframe['filtered_text'] = ''

for index, row in combined_dataframe.iterrows():
    filtered_comment = ' '.join(pattern.findall(row['text']))
    combined_dataframe.at[index, 'filtered_text'] = filtered_comment

# Now, the 'filtered_text' column in the DataFrame contains the text with only English and Georgian letters.
print(combined_dataframe)
```
